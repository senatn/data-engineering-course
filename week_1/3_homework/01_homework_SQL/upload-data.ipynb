```python
gt_iter = pd.read_csv('green_tripdata_2019-01.csv', iterator=True, chunksize=100000)
```

```python
%time gt.to_sql(name='green_tripdata', con=engine, if_exists='append')
```

```python
from time import time

while True:
    try:
        t_start = time()

        gt = next(gt_iter)

        gt.lpep_pickup_datetime = pd.to_datetime(gt.lpep_pickup_datetime)
        gt.lpep_dropoff_datetime = pd.to_datetime(gt.lpep_dropoff_datetime)

        gt.to_sql(name='green_tripdata', con=engine, if_exists='append')

        t_end = time()

        print('inserted another chunk, took %.3f second' % (t_end - t_start))
    except StopIteration:
        print('completed')
        break
```
